```python
import pandas as pd
from nsepy import get_history
from datetime import date
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
data = get_history(symbol="SBIN", start=date(2015,1,1), end=date(2017,1,1))
data.head()
```

```python
data.index = pd.to_datetime(data.index)
data.index

data_cl = data[['Close']]
```

```python
plt.figure(figsize=(100,50))
def moving_average(n_weeks):
    n_days=7*n_weeks
    data_cl['mov_mean']=data_cl.rolling(n_days).mean()

data_cl.plot()
moving_average(1)
plt.plot(data_cl['mov_mean'], color='red', label='Rolling Mean')
print(data_cl['mov_mean'])
```

```python
plt.grid=True
rolling_window10 = 10
rolling_window75 = 75
data_cl.plot()
data_cl['rol_10'] = data_cl['Close'].rolling(rolling_window10).mean()
data_cl['rol_75'] = data_cl['Close'].rolling(rolling_window75).mean()
plt.plot(data_cl['rol_10'], color='red', label='Rolling Mean_10')
plt.plot(data_cl['rol_75'], color='cyan', label='Rolling Mean_75')
plt.plot(legend=2)
```

```python
#3.1
#using 1 for positive direction else negative
i=1
data_vol = data[['Volume']]
data['shock_vol']= 0
data['shock_vol_dir']=0
for i in range(1,data.shape[0]):
    vol_trade = (data['Volume'].iloc[i]-data['Volume'].iloc[i-1])
    vol_trade_per=abs(vol_trade)*100/data['Volume'].iloc[i-1]
    #print(vol_trade_per >=10)
    if vol_trade_per >=  10:
        data.shock_vol.iloc[i]=1
    if vol_trade >=  0:
        data.shock_vol_dir.iloc[i]=1
    
```

```python
#3.2
i=1
data['shock_price']= 0
data['shock_price_dir']=0
for i in range(1,data.shape[0]):
    diff = (data['Close'].iloc[i]-data['Close'].iloc[i-1])
    diff_per=abs(diff)*100/data['Close'].iloc[i-1]
    #print(diff_per)
    if diff_per >=  2:
        data.shock_price.iloc[i]=1
    if diff >=  0:
        data.shock_price_dir.iloc[i]=1
```

```python

no_shock_vol = data[data['shock_vol']==0]
price_shock_without_vol_shock = no_shock_vol[no_shock_vol['shock_price']==1]
data['price_shock_without_vol_shock']=0
data['price_shock_without_vol_shock'] =[1 if (data['shock_price'].iloc[i]==1 and data['shock_vol'].iloc[i]==0) else 0 for i in range(data.shape[0])]
```

```python
from bokeh.plotting import figure
from bokeh.io import output,show
fg = figure(x_axis_type='datetime',x_axis_label='Date',y_axis_label='Timeseries',plot_width=800,plot_height=500)
fg.line(data.index,data['Close'],legend='Close')
i=1

while(i<data.shape[0]):
    start=i
    if(data['shock_vol'].iloc[i]==1 and i+1<data.shape[0] and data['shock_vol'].iloc[i+1]!=1):
        i=i+1
        while(data['shock_vol'].iloc[i]==0 and i<data.shape[0]):
            i=i+1
        end=i
        temp = data.iloc[start:end]
        fg.line(temp.index,temp['Close'],line_color='red',legend='Timeseries between two volume shocks')
    i=i+1
#print(temp)  
fg.circle(price_shock_without_vol_shock.index,price_shock_without_vol_shock['Close'],line_color='red',legend='price shock without vol shock')

```
